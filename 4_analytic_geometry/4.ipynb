**6.**<br>
Выяснить тип кривых второго порядка, порожденных следующими уравнениями:<br>

<p style='background-color:#edf6f9; padding:5px'>Слово инвариантный значит неизменный. 
Инвариантами кривой называются такие выражения, 
составленные из коэффициентов ее уравнения, 
которые не меняются при переходе от одной прямоугольной 
декартовой системы координат к другой такой же системе, 
т.е. при поворотах осей координат 
и при параллельных переносах осей.<br>
(с) "Линейная алгебра и некоторые ее приложения" Л.И. Головина<br><br>
Для кривой второго порядка сумма коэффициентов при квадратах координат $\tau = a_{11}+a_{22}$,<br>
определитель, составленный из коэфиициентов при старших членах $\delta = \begin{vmatrix}
a_{11} & a_{12}\\
a_{12} & a_{22}
\end{vmatrix}$, <br> и определитель третьего порядка $ \Delta = \begin{vmatrix}
a_{11} & a_{12} & a_1 \\
a_{12} & a_{22} & a_2 \\
a_1 & a_2 & a_0 \\
\end{vmatrix} $ являются инвариантами.<br>
    
Общее уравнение кривой второго порядка: <br>
$a_{11}x^2+a_{22}y^2+2a_{12}xy+2a_1x+2a_2y+a_0=0$

<p style='background-color:#edf6f9; padding:5px'>a) $y^2-2x-2y-5=0$<br>

```python
plt.figure(figsize=(12, 5))
y = np.arange(-5, 5, 0.1)
x = (2*y - y**2 + 5)/-2
plt.plot(x, y);
plt.xlim(-4, 4)
plt.title('$y^2-2x-2y-5=0$')

plt.grid(linestyle='--')
```

Выглядит как парабола. Можно проверить с помощью инвариантов кривой второго порядка. 

Коэффициенты: <br>
$a_{11} = 0 \quad a_{22} = 1 \quad a_{12} = 0 \quad 
a_{1} = -1 \quad a_{2} = -1 \quad a_{0} = -5$

$\tau = a_{11} + a_{22} = 1$<br>

$\delta = \begin{vmatrix}
a_{11} & a_{12}\\
a_{12} & a_{22}
\end{vmatrix} = \begin{vmatrix}
0 & 0\\
0 & 1
\end{vmatrix} = 0$ <br>

$ \Delta = \begin{vmatrix}
a_{11} & a_{12} & a_1 \\
a_{12} & a_{22} & a_2 \\
a_1 & a_2 & a_0 \\
\end{vmatrix} = \begin{vmatrix}
0 & 0 & -1 \\
0 & 1 & -1 \\
-1 & -1 & -5 \\
\end{vmatrix} = (-1) \cdot \begin{vmatrix}
0 & 1 \\
-1 & -1 \\
\end{vmatrix} = -1 $

$\delta = 0 \space и \space \Delta \ne 0 \Rightarrow \text{Уравнение } y^2-2x-2y-5=0 \text{ задаёт параболу}.$

<p style='background-color:#edf6f9; padding:5px'>б) $3x^2 + 5y^2 + 12x - 30y + 42 = 0$<br>

```python
plt.figure(figsize=(12, 5))
y = np.arange(-5, 5, 0.1)
x = np.arange(-5, 5, 0.1)

x, y = np.meshgrid(x, y)
plt.contour(x, y,3*x**2 + 5*y**2 + 12*x - 30*y + 42, [1])

plt.xlim(-5, 1)
plt.ylim(-2, 7.5)
plt.title('$3x^2 + 5y^2 + 12x - 30y + 42 = 0$')
plt.grid(linestyle='--')
```

Коэффициенты: <br>
$a_{11} = 3 \quad a_{22} = 5 \quad a_{12} = 0 \quad 
a_{1} = 6 \quad a_{2} = -15 \quad a_{0} = 42$

$\tau = a_{11} + a_{22} = 8$<br>

$\delta = \begin{vmatrix}
a_{11} & a_{12}\\
a_{12} & a_{22}
\end{vmatrix} = \begin{vmatrix}
3 & 0\\
0 & 5
\end{vmatrix} = 15$ <br>

$ \Delta = \begin{vmatrix}
a_{11} & a_{12} & a_1 \\
a_{12} & a_{22} & a_2 \\
a_1 & a_2 & a_0 \\
\end{vmatrix} = \begin{vmatrix}
3 & 0 & 6 \\
0 & 5 & -15 \\
6 & -15 & 42 \\
\end{vmatrix} = 3 \cdot \begin{vmatrix}
5 & -15 \\
-15 & 42 \\
\end{vmatrix} + 6 \cdot \begin{vmatrix}
0 & 5 \\
6 & -15 \\
\end{vmatrix}  = -225 $

$\delta > 0, \Delta \ne 0 \text{ и } \tau \cdot \Delta < 0 \Rightarrow \text{Уравнение } 3x^2 + 5y^2 + 12x - 30y + 42 = 0 \text{ задает эллипс}$.

<p style='background-color:#edf6f9; padding:5px'>в) $ 2x^2-y^2 + 6y - 7 = 0$<br>

```python
plt.figure(figsize=(12, 5))
y = np.arange(-10, 10, 0.1)
x = np.arange(-10, 10, 0.1)

x, y = np.meshgrid(x, y)
plt.contour(x, y,2*x**2-y**2+6*y-7, [1])

plt.ylim(-3, 10)
plt.title('$ 2x^2-y^2 + 6y - 7 = 0$')
plt.grid(linestyle='--')
```

Коэффициенты: <br>
$a_{11} = 2 \quad a_{22} = -1 \quad a_{12} = 0 \quad 
a_{1} = 0 \quad a_{2} = 3 \quad a_{0} = -7$

$\tau = a_{11} + a_{22} = 1$<br>

$\delta = \begin{vmatrix}
a_{11} & a_{12}\\
a_{12} & a_{22}
\end{vmatrix} = \begin{vmatrix}
2 & 0\\
0 & -1
\end{vmatrix} = -2$ <br>

$ \Delta = \begin{vmatrix}
a_{11} & a_{12} & a_1 \\
a_{12} & a_{22} & a_2 \\
a_1 & a_2 & a_0 \\
\end{vmatrix} = \begin{vmatrix}
2 & 0 & 0 \\
0 & -1 & 3 \\
0 & 3 & -7 \\
\end{vmatrix} = 2 \cdot \begin{vmatrix}
-1 & 3 \\
3 & -7 \\
\end{vmatrix} = -4 $

$\delta < 0, \Delta \ne 0 \Rightarrow \text{Уравнение } 2x^2-y^2 + 6y - 7 = 0 \text{ задает гиперболу}$.

<p style='background-color:#edf6f9; padding:5px'>г) $ 2x^2 - 3y^2 - 28x - 42y - 55 = 0$

```python
plt.figure(figsize=(12, 5))
y = np.arange(-100, 100, 0.1)
x = np.arange(-100, 100, 0.1)

x, y = np.meshgrid(x, y)
plt.contour(x, y,2*x**2-3*y**2-28*x-42*y-55, [1])

# plt.xlim(-5, 1)
# plt.ylim(-3, 10)
plt.title('$ 2x^2 - 3y^2 - 28x - 42y - 55 = 0$')
plt.grid(linestyle='--')
```

Коэффициенты: <br>
$ a_{11} = 2 \quad a_{22} = -3 \quad a_{12} = 0 \quad 
a_{1} = -14 \quad a_{2} = -21 \quad a_{0} = -55 $

$\tau = a_{11} + a_{22} = -1$<br>

$\delta = \begin{vmatrix}
a_{11} & a_{12}\\
a_{12} & a_{22}
\end{vmatrix} = \begin{vmatrix}
2 & 0\\
0 & -3
\end{vmatrix} = -6$ <br>

$ \Delta = \begin{vmatrix}
a_{11} & a_{12} & a_1 \\
a_{12} & a_{22} & a_2 \\
a_1 & a_2 & a_0 \\
\end{vmatrix} = \begin{vmatrix}
2 & 0 & -14 \\
0 & -3 & -21 \\
-14 & -21 & -55 \\
\end{vmatrix} = 2 \cdot \begin{vmatrix}
-3 & -21 \\
-21 & -55 \\
\end{vmatrix} - 14 \cdot \begin{vmatrix}
0 & -3 \\
-14 & -21 \\
\end{vmatrix} = 36 $

$\delta < 0, \Delta \ne 0 \Rightarrow \text{Уравнение } 2x^2 - 3y^2 - 28x - 42y - 55 = 0 \text{ задает гиперболу}$.
