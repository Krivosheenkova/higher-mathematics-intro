**4.** Постройте на одном графике две кривые y(x) для функции двух переменных y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.


```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(16,8))
k = 1
x = np.arange(-20, 20, 0.1)
y = [np.cos(k*_) for _ in x]
plt.plot(x, y, label='k=1, y(k, x) = cos(x)');

k = 10
x = np.arange(-20, 20, 0.1)
y = [np.cos(k*_) for _ in x]
plt.plot(x, y, label=f'k={k}, y(k, x) = cos({k}x)');

plt.xlabel('x')
plt.ylabel('y');
plt.legend();
```
